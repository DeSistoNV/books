# Mask Thesholding Visualization

```python
import nibabel as nib
import numpy as np
from nilearn.plotting import plot_glass_brain
%pylab inline
```

```python
mask = nib.load('/home/nick/Downloads/white.nii')
```

# Original Mask

```python
plot_glass_brain(mask,colorbar=True)
```

```python
data = np.asarray(mask.get_data())
```

# Mask at different thresholds (binary)
* NOTE : voxels selection number here will differ from number of voxels selected from DTI data, as the mask has not yet been resampled.

```python
thresholds = np.arange(0,1,.1)
for i,t in enumerate(thresholds):
    d = np.copy(data)
    d[d < t], d[d > t] = 0 , 1
    plot_glass_brain(nib.Nifti1Image(d,affine=mask.get_affine()),
                                     vmax=1,
                                     vmin=0,
                                     title='threshold : {} voxels selected : {}'.format(t,sum(d == 1))
                                    )
```
